# Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from removeFrame import removeFrame
```

# Folding rule example

```python
image_path = 'tummstock/IMG_0159.jpg'
output = removeFrame(image_path)

if output is not None:
    print(image_path)
    fig, (ax1,ax2) = plt.subplots(1,2, figsize=(6, 6), facecolor='w', edgecolor='k')
    ax1.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
    ax1.set_title('Original')
    ax2.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    ax2.set_title('Without frame')
    plt.show()
```
