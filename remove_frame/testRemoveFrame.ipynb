```python
import cv2
import numpy as np
import os
from removeFrame import removeFrame
from houghLines import removeFrameUsingHoughLines
import matplotlib.pyplot as plt
```

```python
# Use Houghlines implementation on frames which uses a folding rule as frame
for i, image in enumerate(os.listdir(os.getcwd()+'/tummstock')):
    
    img = cv2.imread(os.getcwd()+'/tummstock/'+image)
    
    without_frame = removeFrameUsingHoughLines(os.getcwd()+'/tummstock/'+image)
    
    plt.imshow(without_frame)
    plt.xlabel(image)
    plt.title('Image without frame')
    plt.show()
    if i == 9:
        break
```

```python
# Use implementation on frames which uses a blue frame
for i, image in enumerate(os.listdir(os.getcwd()+'/blue')):
    
    img = cv2.imread(os.getcwd()+'/blue/'+image)
    
    without_frame = removeFrame(os.getcwd()+'/blue/'+image)
    
    plt.imshow(without_frame)
    plt.xlabel(image)
    plt.title('Image without frame')
    plt.show()
    
    if i == 9:
        break
```
